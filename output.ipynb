```python
import matplotlib.pyplot as plt
```

```python
loss = []
policy_loss = []
eprewmean = []
with open('output.txt', 'r') as f:
    lines = f.readlines()
    for l in lines:
        arr = l.split('|')
        if len(arr)>=3 and arr[1].strip()=='value_loss':
            loss.append(arr[2].split())
        if len(arr)>=3 and arr[1].strip()=='policy_loss':
            policy_loss.append(arr[2].split())
        if len(arr)>=3 and arr[1].strip()=='eprewmean':
            mean = arr[2].strip().split('e')
            eprewmean.append(float(mean[0])*(10**int(mean[1])))
#             print(int(arr[2].strip()))
#             eprewmean.append(arr[2].split())
```

```python
plt.plot([i for i in range(len(loss))], [float(i[0]) for i in loss])
plt.xlabel('time_step',size = 18)
plt.ylabel('value_loss', size = 18)
plt.title('value_loss - time step', size = 18)
plt.savefig('value_loss', dpi = 200)
```

```python
plt.plot([i for i in range(len(policy_loss))], [float(i[0]) for i in policy_loss])
plt.xlabel('time_step',size = 18)
plt.ylabel('policy_loss',size = 18)
plt.title('policy_loss - time step',size = 18)
plt.savefig('policy_loss', dpi = 200)
```

```python
plt.plot([i for i in range(len(eprewmean))], [i for i in eprewmean])
plt.xlabel('time_step',size = 18)
plt.ylabel('mean_reward',size = 18)
plt.title('mean_reward - time step',size = 18)
plt.savefig('mean_reward', dpi = 200)
```
